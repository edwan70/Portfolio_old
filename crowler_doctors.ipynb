```python
url = 'https://prodoctorov.ru/novorossiysk/stomatolog/?page=1'
data = []
data1 = []

for page in range(1, 12):
    url = f'https://prodoctorov.ru/novorossiysk/stomatolog/?page={page}'
    r = requests.get(url)
    print(url)
    #time.sleep(1)
    soup = BeautifulSoup(r.text, 'html.parser')
    doctors = soup.find_all('div', class_='b-doctor-card__main-container')
    address_all = soup.find_all('div', class_='b-doctor-card__appointment-container')
    for doctor in doctors:
        fio = doctor.find('span' , class_='b-doctor-card__name-surname').text.strip()
        spec = doctor.find('div' , class_='b-doctor-card__spec').text.strip().split()
        spec = ' '.join(spec)
        link = 'https://prodoctorov.ru'+doctor.find('a' , class_='b-doctor-card__name-link').get('href')
        url1 = link
        req = requests.get(url1)
        print(url1)
        time.sleep(3)
        soup1 = BeautifulSoup(req.text, 'html.parser')        
        rating = soup1.find('div', class_='ui-text ui-text_h5 ui-text_color_black font-weight-medium mr-2')\
        .text.strip()
        
        try:
            exp = doctor.find('div' , class_='b-doctor-card__experience-years').text.strip()
        except:
            exp ='-'
        data.append ([fio, spec ,link , exp, rating])
    for adress in address_all:
        address = adress.find('div' , class_='b-doctor-card__select-lpu-container').text.strip().split()
        address = ' '.join(address)
        data1.append(address)
print('Конец цикла перебора страниц. Конец работы кода.')
```
